```python
#Generator Parameter
g = 1/16
#Number of bits for left shift (used to circumvent floating point calculations)
precision_bits = 7
#number of steps to print
reduction_steps = 29
```

```python
a = 1.0
j = 1
prec = 2**precision_bits
str_list = ""
for i in range(0,reduction_steps):
    str_list += str(round(prec *(1/(1-(a/2))))) + ", "
    a = (1-g)*a

print('[' + str_list + ']')
#remove the last comma :)
```
